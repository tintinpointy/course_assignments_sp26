```python
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np

files = ["M13_0001.fits", "M13_0002.fits", "M13_0003.fits", "M13_0004.fits", "M13_0005.fits"]

fits_data = []

for path in files:
    with fits.open(path) as hdul:
        fits_data.append(hdul[0].data)
```

```python
image = np.median(np.array(fits_data), axis = 0)
print(image)
```

```python
plt.hist(image.ravel(), bins = 50)
plt.show()
```

```python
v_min = 300
v_max = 400

plt.imshow(image, origin = "lower",  vmin = v_min, vmax = v_max)
plt.colorbar()
plt.show()
fits.writeto('combined.fits', image, overwrite = True)
```
