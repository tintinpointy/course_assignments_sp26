```python
import numpy as np

my_list = [20,27,9,12,31,17,4,50,39,35]
arr = np.array(my_list)
print("Multiply by 4:", arr*4)
```

```python
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
min_index = np.argmin(arr)
max_index = np.argmax(arr)
print(f"Index of smallest: {min_index}, Index of largest: {max_index}")
```

```python
min_val = np.min(arr)
max_val = np.max(arr)
print(f"Smallest value (np.min): {min_val}, Largest value (np.max): {max_val}")
```

```python
#Yes they do, at Index 6 the smallest value is there (4), and at Index 7 the largest (50).
more20 = arr[arr > 20]
print(more20)
```

```python
matrix = np.array([[5,12,17,9],[8,21,14,3],[6,13,19,11]])
print(matrix.shape)
```

```python
#3 rows, 4 columns
print(matrix[:, 2])
print(matrix[1, :])
print(matrix*2)
```

```python
print(np.mean(matrix, axis=0))
print(np.mean(matrix))
print(np.max(matrix, axis=1))
print(np.min(matrix, axis=0))
flat_mat = matrix.flatten()
print(flat_mat)
print(flat_mat.shape)
```

```python
#A line -- the matrix was flattened into a single array of twelve elements
flat_min_index = np.argmin(matrix)
flat_max_index = np.argmax(matrix)
print(flat_min_index)
print(flat_max_index)
print(np.unravel_index(flat_min_index, matrix.shape))
print(np.unravel_index(flat_max_index, matrix.shape))
```

```python
more15 = matrix[matrix > 15]
print(more15)
```

```python
sorted_matrix = np.sort(matrix, axis = 1)
print(sorted_matrix)
```

```python
arr = np.array([[2,3,5],[14,6,8],[11,13,17],[7,10,13]])
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(np.sqrt(num)) + 1):
        if num % i ==0:
            return False
    return True

prime_vectors = np.vectorize(is_prime)

def get_prime_rows(arr):
    mask = prime_vectors(arr)
    prime_rows = np.any(mask, axis = 1)
    return arr[prime_rows]

print(get_prime_rows(arr))
```

```python
def checkerboard():
    board = np.zeros((8,8), dtype = int)
    board[0::2, 0::2] = 1
    board[1::2, 1::2] = 1
    return board
print(checkerboard())
```

```python
def reverse_checkerboard():
    board = np.zeros((8,8), dtype = int)
    board[0::2, 1::2] = 1
    board[1::2, 0::2] = 1
    return board
print(reverse_checkerboard())
```

```python
def expand(arr, num_spaces):
    spacer = " " * num_spaces
    def expand_word(word):
        return spacer.join(list(word))
    vector_expand = np.vectorize(expand_word)
    return vector_expand(arr)
universe = np.array(['galaxy','clusters'])
expand(universe,1)
expand(universe,2)
```

```python
def secondBrightest(star):
    sorted_star = np.sort(star,axis = 1)
    return sorted_star[:, 1]

np.random.seed(123)
star = np.random.randint(500, 2000, (5,5))
print(star)
print(secondBrightest(star))
```
